## REPLICATION SET 

1. **Replication set** is next gen replication controller
2. It supports a new selector that can do selection based on filtering according to a set of values
    1. e.g "environment" either "dev" or "qa"
    2. not only based on equality like Replication Controller
        1. e.g. "environment" =="dev"
3. This **Replica Set**, rather than the Replication Controller, is used by Deployment object.

### DEPLOYMENT 

1. A deployment declaration in Kubernetes allows you to do app **deployments** and **updates**
2. When using deployment object, you define the **state** of your application.
    1. Kubernetes will then make sure the cluster matches **desired** state
3. Just using **Replication Controller*8 or **Replication Set** might be **cumbersome** to deploy apps.
    1. **Deployment object** is easier to use and gives you more possibilities.
