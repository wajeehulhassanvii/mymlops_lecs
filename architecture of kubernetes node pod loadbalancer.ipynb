Any pod can have multiple containers, in diagram above pod1 has 3, pod2 has 1, pod5 has 1, pod4 has 3.

Any node can have multiple pods

containers within pods can can use localhost and port number to communicate with each other

nodes are in cluster, and pods can communicate with another pod but it will have to go over the network

**kubelet** is responsible for launching the pod, it connect to master node and get the info

**kube-proxy** is going to feed information about what pods are on this node to iptable, iptable in linux is firewall and can route traffic. Whenever a pod is launched then kubeproxy changes the iptable to make sure the pod become routable

**load-balancer** is a public service, client connect to load balancer and then load balancer forward traffic to kubernetes cluster. I t will have all the nodes listed and the traffic will arrive to iptable, iptable has the rules to forward traffic to another node. if the pod is in different node then iptable will send the traffic to another node

we expose nodeport static, comm b/w service and nodes is going over one port

in aws, aws elb has one exposed port which routes all the traffic to node and  iptables
