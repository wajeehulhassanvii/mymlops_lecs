##### we will deploy below docker file

the app is a webservice and we are copying some files in the container

we are creating our own docker image so we will first do<br>
**docker build .**

we can use CI/CD software like Jenkins/Travis to automate the build process

we can run Docker app by running **Docker run** locally

#### to make the image available to Kubernetes, you need to push the image to Docker Registry, like Docker HUB 

1. make an account on Docker Hub
2. push any local built image to Docker Registry

##### steps for pushing image to Docker Registry
1. docker login
2. docker tag (imageid) (your-login)/(docker-demo)
3. docker push (your-login)/(docker-demo)

##### we can tag docker image during build by running the following command:
**docker build -t (your-login)/(docker-demo)** <br>
**docker push (your-login)/(docker-demo)**

##### LIMITATIONS:
1. run one process in one container
2. Don't create one giant docker image for the app, but split it up if necessary
3. All the data in the container is not presernved, as the container stops, the data get lost
4. Data can be preserved using volumes
