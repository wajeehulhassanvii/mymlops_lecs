##### docker run = docker create + docker start

during creation it pulls the file system snapshot and place it in container, then run runs the command we give it like 'ls'

**docker create hello-world**<br> creates a docker a image and gives us random string<br>now we can paste that random container id string to run it<br>
**docker start (random container id string)**<br>
without a it wont print on terminal there will be no watch, so we add -a for watching output
**docker start -a (random container id string)**

if docker container has exited we can start it with its string to the following command:<br>
**docker start stringId**

## run redis 

##### docker run redis <br>
it runs a redis container, which we can acess from another terminal using below command

##### docker exec -it  (container id) redis-cli
to get redis-cli running onto our already running container

-it is two seperate flags, -i is for input and -t is for terminal for STDIN and STDERR

##### docker exec -it (container id) shell|zsh|sh <br>
runs the shell of the running container

##### docker run -it busybox sh
will only run sh from busybox, not other processes will be running
