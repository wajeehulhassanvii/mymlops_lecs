like livenessProbe, indicates whether a container is running
- if check fials then container will be restarted in **livenessProbes**
- **readinessProbe** tells whether the container is ready to serve requests
<br><br>
if check fails, the container **will not be restarted**, but **the Pod's IP address will be removed from the service**, so itll not serve any request anymore

- will make sure at startup that the pod only recieve traffic when the test succeeds
- you can use these probes **in conjunction**, and you can configure different tests for them

in general we configure both the livenessProbe and readinessProbe

livenessProbe and ReadinessProbe both can be defined simultaneously
