##### Before we deploy we must create a **pod definition** first 

a pod can contain one or more tightly coupled containers, that make the app
1. those apps can easily communicate with each other using their local port numbers

we can use **kubectl** to create the pod on kubernetes cluster

to create a cluster from pod configuration, we run the below command:
<br>**kubectl create -f k8s-demo/pod-helloworld.yml**

## USEFUL POD COMMANDS 

## we will deploy below code on kubernetes cluster 

git clone https://github.com/wardviaene/kubernetes-course

##### first check if kubectl is running by issuing the command:
**kubectl get node** , it will igive us at least one minikube node

check the pod configuration, cat first-app/helloworld.yml

then we run **kubectl create -f first-app/helloworld.yml**

then we check pods by running **kubectl get pod**

##### to get details about the pod:
run the command **kubectl describe pod (pod name from kubectl get pod)**

## kubectl describe pod will get the pod running, to get to the pod we can use port forward technique 

##### kubectl port-forward nodehelloworld.example.com 8081:3000

the above command will forward and attach the pod port 3000 to localhost 8081, and we can test the pod by running **curl localhost:8081**

### second way is by creating a service 

**kubectl expose pod nodehelloworld.example.com --type=NodePort --name nodehelloworld-service**<br>by running this command we can directly access the kubernetes 
pod by port 3000

##### to get the ip of the pod, we can run the command:
**minikube service nodehelloworld-service --url**

**kubectl get service** command will give us the internal IP address and other informations

## more commands

**kubectl attach nodehelloworld.example.com** will attach to the running process, we can see log or we can see nothing depends on the process that is running in the pod.

**kubectl exec nodehelloworld.example.com -- ls /app** will execute ls command in the running pod

**kubectl get service** 

**kubectl describe service nodehelloworld-service** will give namespace, label, selector, type, ip, port, nodeport, endpoints, session affinity

##### we can launch another pod busybox and connect to our kubernetes cluster from there 

kubectl run -i --tty busybox --image=busybox --restart=N
