On real world we need to be able to access app from outside the cluster.<br>
On AWS, you can easily add an external load balancer
<br>This AWS Load Balancer will route the traffic to the correct pod in kubernetes
<br>There are other solutions for other cloud providers that don't have a load balancer
<br> You can have **haproxy/Nginx** load balancer in front of our cluster<br>
we can expose directly as well

## RUNNING POD AND SERVICE (LOAD BALANCER) 

in service file , the targetPort is set to the name of the Pod->port:name, which is "nodejs-port" in the example,<br>
**Selector** is selecting the **Pod** for which the current service file is defined

1. first we run **kubectly create -f first-app/helloworld.yml**, to run the cluster
2. then we run the service in the same manner **kubectl create -f first-app/helloworld-service.yml**

### note: since we have our aws configured, the cluster will be created in amazon cloud service and we can check newly created services like load balancer in the console 

then we can create a dns namespace for load balancer, so load balancer can handle al; the requests
<br>
1. now when we send request, it will go  to load balancer
2. then the load balancer will redirect the request to one of those namespaces

kubectl will itself clreate all the configurations and security configurations in AWS

so that all the pods can communicate with each other
