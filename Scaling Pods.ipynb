# SCALING  

if your application is **stateless** you can do horizontal scaling
    1. Stateless means that the application does not have any state and does not write any local file and neither does it keeps local session.

if we have two pods and one pod is going to write something lcoally then those pods will be out of sync, and we won't be able to scale it

Traditional databases like MySQL, Postgres are **stateful**, they have database files that can't be split over multiple instances

most web apps can be made stateless
    1. session management needs to be done outside the containers, we can use Redis, database or something else
    2. anny file that needs to be saved **can't be saved locally** on the container, in AWS we can use S3 bucket as **volume**

**Volumes** allow us to run stateful apps<br>
**Stateful apps** can't horizontally scale but you can run them in a single container and scale **vertical** (allocate more CPU / Memory / Disk)

## REPLICATION CONTROLLER 
1. Scaling in Kubernetes can be done using the **Replication controller**
2. The replication controller will ensure a specified number of **pod replicas** will run at all time
3. A pod created with the replica controller will automatically be replaced if they fail, get deleted or are terminated
4. Using replication controller is recommended if we just want to make sure that 1 pod is always running, even after reboots
    1. You can then run a replication controller with just 1 replica
    2. This makes sure that the pod is always running
