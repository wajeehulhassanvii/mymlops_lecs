kubeflow sdk and DSL domain specific language will be used for creating kubeflow pipeline.

- for pipeline we can define a function and decorate it with dsl.pipeline
    - the function will have parameters, which we can change from kubeflow dashboard as well
    - inside func we can define our steps

1. In first step, we can have our hyper-parameter tuning, we will use Katib component and we will define objective, metrics, algorithm, and parameters that need to be tuned alongwith the visible space
2. Then in raw template, we specify the specsb by providing docker image, name and command
3. then we take raw template and convert into JSON format.
4. then we can use reusable KATIB component available on github page
5. next we can use set parameters for components with values we created
6. then we will create a small comp, which will echo/print optimized result from katib hyperparameter
7. **Then we will pass the optimized output to katib.op.output**
    - Thats how we passed output from one step of pipeline to another
    - also automatically define the dependency between the steps
8. finally we compile the pipeline as a tar in main function using kfp.compiler

We will then get the compiled version of the workflow, which we will upload to the kubeflow pipeline.

we can create experiment after uploading the pipeline.
