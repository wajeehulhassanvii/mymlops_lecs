## deploying kubernetes cluster to aws using kops 

kops require aws to be configured, then run the below command

kops create cluster --name=kubernetes.dodgecorona.com --state=s3://(aws bucket name) --zone=as-east-1a --node-size=t2.micro -- master-size=t2.micro --dns-zone=kubernetes.dodgecorona.com

above command is self explanatory, we made 2 nodes, one as node and another one as master

after running the command we can see the configurations

##### then to launch the cluster just run the following command
kops update cluster kubernetes.dodgecorona.com --yes

##### otherwise we can make changes to cluster beofre its deplyed by running the following command 

kops edit cluster kubernetes.dodgecorona.com --state=s3://kops-state-b429try

##### kops will update the ~/.kube/config file after the cluster has been configured 

#### kubectl will be able to get the external node information as well, we can do that by running
kubectl get node

### we only configured kubernetes cluster on aws, now we will do deployment of a test cluster 

kubectl run hello-minikube --image=k8s.gcr.io/echoserver:1.4 --port=8080

##### now we can expose the deployment 

kubectl expose deployment hello-minikube --type=NodePort

we canconnect to any port except master

##### to delete the cluster just run
kubectl delete cluster kubernetes.dodgecorona.com
