we will launch the below configuration file into production

Launch the kubernetes cluster by running the following command<br>
**kubectl create -f kubernetes-course/replication-controller/helloworld-repl-controller.yml**

then we can get pod informations by issuing the command:
**kubectl get pods**

we can see that replication controller created two pods, hello-world controller-79qlq and hello-kubernetes

##### we can describe the pods now 

##### kubectl describe  pod helloworld-controller-8qfsw//////helloworld-controller-x9lg9

we can see that we have multiple pods running and we have scaled the application horizontally

**if we had multiple instances running on multiple nodes, we could put load balancer tomake multiple ports accessible to other software or customers**
If one of those underlying node crashes then the controller will automatically re-schedule the pods

if we delete one of the pod then replication manager will create a new pod itself

### we can scale those pods by explicitelky specifying "--replicas=X", issue the below "scale" command, we are using FILENAME here

**kubectl scale --replicas=4 -f kubernetes-course/replication-controller/helloworld-repl-controller.yml**

## get the info of deplyed replicas by issuing below command 

##### kubectl get rc 

#### scale with REPLICA CONTROLLER NAME 

kubectl scale --replicas=1 -f rc/helloworld-controller

we can only scale horizontally if the pod is stateless, if stateful then we won't be able to do it
